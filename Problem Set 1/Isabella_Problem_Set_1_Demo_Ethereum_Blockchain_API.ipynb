### Question 3: 

What are the changes of Ethereum before and after the merge? Please identify at least three different changes to elaborate on. 

Please provide references that could back-up your answer. 

Please insert Figures/Tables to better represent your idea. 





"The merge" combined the original Ethereum platform, a proof-of work execution layer, with the Beacon chain, a proof-of-stake consensus layer [(The Merge 2022)](https://ethereum.org/en/upgrades/merge/). The proof-of-work system was removed and replaced by the proof-of-stake system, which reduced energy consumption and improved scalability for the future [(The Merge 2022)](https://ethereum.org/en/upgrades/merge/).

the merge image.jpg

[(The Merge 2022)](https://ethereum.org/en/upgrades/merge/)

The merge eliminated the need for Ethereum mining, lowering energy consumpution [(Nambiampurath 2022)](https://www.investopedia.com/ethereum-completes-the-merge-6666337). After the merge, Ethereum's energy consumption dropped by about 99.95% [("The Merge" 2022)](https://ethereum.org/en/upgrades/merge/). As shown in the image below, their energy consumption (both PoW and PoS) compared to other to companies like Bitcoin and YouTube are much lower [(Ethereum Energy Consumption 2022)](https://ethereum.org/en/energy-consumption/).

ethereum energy expenses.jpg

[(Ethereum Energy Consumption 2022)](https://ethereum.org/en/energy-consumption/).

Finally, the merge also impacted how ETH is issued. Before the merge, about 13,000 ETH were issued daily through mining and staking rewards. Now, mining has been eliminated and only about 1,600 ETH are issued daily in staking rewards [(The Merge Ethereum Upgrade: All You Need to Know 2022)](https://academy.binance.com/en/articles/the-merge-ethereum-upgrade-all-you-need-to-know). Ethereum hopes that by reducing the issuance of ETH by almost 90%, the market value of ETH will rise [(The Merge Ethereum Upgrade: All You Need to Know 2022)](https://academy.binance.com/en/articles/the-merge-ethereum-upgrade-all-you-need-to-know).
